```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_excel(next(iter(uploaded)))
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Origem', order=df['Origem'].value_counts().index)
plt.title('Leads por Origem')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df, x='Etapa do Funil', order=df['Etapa do Funil'].value_counts().index)
plt.title('Leads por Etapa do Funil')
plt.xticks(rotation=45)
plt.show()
```

```python
conversao = df['Virou Cliente'].value_counts(normalize=True) * 100
plt.figure(figsize=(6,6))
conversao.plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'salmon'])
plt.title('Taxa de Conversão de Leads em Clientes')
plt.ylabel('')
plt.show()
```
